```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("output.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df
```

```python
df.columns
```

```python
y = df['Target variable ( 1: Defaulters / 0: Non-Defaulters)']
df['Target variable ( 1: Defaulters / 0: Non-Defaulters)'].value_counts()
```

```python
#so there are two different categories (default vs. non-default)

#do not use linear regression, this is a classification problem
#start with random trees classification to get useful variables
df = df.drop(columns="Customer ID")
df.columns
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df.drop(columns='Target variable ( 1: Defaulters / 0: Non-Defaulters)')
selected_features = selected_features.drop(columns='Dealer codes from where customer has purchased the Two wheeler')
selected_features.columns
```

# Create a Train Test Split

Use `Target variable ( 1: Defaulters / 0: Non-Defaulters)` for the y values

```python
from sklearn.model_selection import train_test_split

X = selected_features

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, train_size=0.8, random_state=22)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data

#setup MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

#set initial sizing of minmaxscaler using training data
X_scaler = MinMaxScaler().fit(X_train)

#y is just 1 or zero, so no need to scale it!
# y_scaler = MinMaxScaler().fit(y_train)
```

```python
#use MinMaxScaler to transform the data in the training and test data sets
#so that it is not weighted improperly

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
# y_train_scaled = y_scaler.transform(y_train)
# y_test_scaled = y_scaler.transform(y_test)
```

# Train the Model



```python
#first, use random forest classification in this notebook (to get best features to use later)!
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators=200)
classifier
```

```python
#create the model using the training data!
classifier.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
importances = classifier.feature_importances_
sorted(zip(importances, selected_features.columns), reverse=True)
```
